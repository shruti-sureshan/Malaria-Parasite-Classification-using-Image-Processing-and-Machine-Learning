```python
from os import listdir, path
from math import floor
from skimage import io
import cv2
import numpy as np
#malaria_dataset_path = r'C:\Users\USER\Downloads\cell_images'
parasitized_dir = r'C:\Users\USER\Downloads\cell_images\Parasitized'
uninfected_dir = r'C:\Users\USER\Downloads\cell_images\Uninfected'
```

```python
import csv
def getbins(cell_images_path):
    cell_image_files = listdir(cell_images_path)
    cell_image_files.sort()
    num_cell_images = len(cell_image_files)
    data = np.zeros((num_cell_images,50,50,3))
    with open('C:\Users\SHRUTI SURESHAN\Desktop\image_bins_all.csv', 'w', newline='') as f:
        hey = csv.writer(f)
        hey.writerow(['Bins0','Bins1','Bins2','Bins3','Bins4','Bins5','Bins6','Bins7'])
        for cell_image_file in cell_image_files:
            cell_image = io.imread(path.join(cell_images_path, cell_image_file))
            a = np.array(cell_image[:,:,0], dtype = np.uint8)
            d = dict()
            for i in range (0,256):
                d[i]=0

            x = []
            for i in range (0,256):
                x.append(i)    

            y = []
            for i in range (0,256):
                y.append(d[i])

            num = 0
            den = 0
            for i in a.flatten():
                d[i] = d.get(i,0) + 1

            b = 0
            for i in x:
                b = d.get(i,0) + 1
                num = num + b*i
                den = den + b
            cogr = num / den
        
            a = np.array(cell_image[:,:,1], dtype = np.uint8)
            d = dict()
            for i in range (0,256):
                d[i]=0

            x = []
            for i in range (0,256):
                x.append(i)    

            y = []
            for i in range (0,256):
                y.append(d[i])

            num = 0
            den = 0
            for i in a.flatten():
                d[i] = d.get(i,0) + 1

            b = 0
            for i in x:
                b = d.get(i,0) + 1
                num = num + b*i
                den = den + b
            cogg = num / den
        
            a = np.array(cell_image[:,:,2], dtype = np.uint8)
            d = dict()
            for i in range (0,256):
                d[i]=0

            x = []
            for i in range (0,256):
                x.append(i)    

            y = []
            for i in range (0,256):
                y.append(d[i])

            num = 0
            den = 0
            for i in a.flatten():
                d[i] = d.get(i,0) + 1

            b = 0
            for i in x:
                b = d.get(i,0) + 1
                num = num + b*i
                den = den + b
            cogb = num / den
        
            h, w, _ = cell_image.shape
            R, G, B = 0, 1, 2

            for color, cog in [(R, cogr), (G, cogg), (B, cogb)]:
                for r in range(h):
                    for c in range(w):
                        if cell_image[r, c, color] <= cog:
                            cell_image[r, c, color] = 0
                        else:
                            cell_image[r, c, color] = 1
        
            bins0 = 0
            bins1 = 0
            bins2 = 0
            bins3 = 0
            bins4 = 0
            bins5 = 0
            bins6 = 0
            bins7 = 0
            for i in range(h):
                for j in range(w):
                    r= cell_image[i,j,0];
                    g= cell_image[i,j,1];
                    b= cell_image[i,j,2];
        
                    if r==0 and g==0 and b==0:
                            bins0= bins0 + 1
                    elif r==0 and g==0 and b==1:
                            bins1= bins1 + 1
                    elif r==0 and g==1 and b==0:
                            bins2= bins2 + 1
                    elif r==0 and g==1 and b==1:
                            bins3= bins3 + 1
                    elif r==1 and g==0 and b==0:
                            bins4= bins4 + 1
                    elif r==1 and g==0 and b==1:
                            bins5= bins5 + 1
                    elif r==1 and g==1 and b==0:
                            bins6= bins6 + 1
                    elif r==1 and g==1 and b==1:
                            bins7= bins7 + 1
            
            hey.writerow([bins0,bins1,bins2,bins3,bins4,bins5,bins6,bins7])
    
    data = data.astype('float32')
    return data
        
```
